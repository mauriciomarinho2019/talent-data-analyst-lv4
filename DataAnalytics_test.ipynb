**Questão 3** - Quais são as camadas de um Data Lake?

```python
Podemos assim classifica-las, as camadas de um data lake podem seguir a seguinte sequecia: 
 	 ==) Data Source: Gerencia de onde vem os dados, fontes como csv, planilhas, tabelas em bancos, streaming de dados (kafka, rabbitMQ) 
    ==) Ingestão: Tráfego dos dados , uma camada de armazenamento de dados brutos sem nenhuma transformação aqui entra um (Apache Flame, Storm)
      ==) Caching: Onde os dados ficam de forma temporaria para os próximos processos  ==) Processamento: tratamento dos dados com aplicações especificas aqui tambem pode ser feito o enriquecimento dos dados e descarte dele para proximas etapas 
        ==) Downstream: Nessa etapa, alguma ferramenta é utilizada para organização dos dados com vinculo com a camada anterior, segmentação dos dados para cada tabela e departamento.
         ==) Delivery: Nessa etapa é a forma como os dados serão consumidos, sejam por apis, BIs ou datasets para times de datascience
```

```python
Principais e suscintas diferenças entre DL e DW:
Data lake aceita uma enorme quantia de dados e diversos tipos, ou seja, trabalha com dados brutos e volumosos.  Podendo ser dados 
de videos, imagens, csvs, planilhas, arquivos de audio, json, quais quer tipos de dados, dessa forma é possivel usar um Data Lake 
para os processos e estruturas que agregem mais valor a empresa, cruzando diversos dados para resolver as principais questões de 
Negocios. 
DL é muito usado para que empresas consigam organizar seus dados a fim de obterem insights para tomada de decisão com mais velocidade e rapidez

DataWarehouse, aceita apenas dados organizados, estruturados,ou seja requintados, gerencia uma menor quantia 
dados em comparação ao data lake que suporta mais dados, porém cada um com suas vantagens e diferenças.
O DW apresenta um ponto excelente para retorno rapido e de facil implantação
```

```python
# Essa funcao remove caracter especial nas colunas(document, phoneNumber, birthdate)
def remove_caracter_especial_desafio2 (var_especial):
    return var_especial.replace ("/","").replace ("+","").replace ("(","").replace (")","").replace (",","").replace (".","").replace ("-","").replace (" ","")

#  Detectar através da coluna `document` se o registro é de uma Pessoa Física ou Pessoa Jurídica, adicionando uma coluna com essa informação 
check_document = lambda document: "pessoa juridica" if len(str(document)) == 14 else "pessoa fisica"
 
#  Higienizar e homogenizar o formato da coluna `birthDate`
limpa_date = lambda date: parse (date)

## resolve colunas com dados em campo errado (colunas trocasas)
def correcao_de_dados_campo (base_pessoas):
    base_auxiliar = base_pessoas.copy ()
    
    # - duplicata = base_antiga copy base_nova
    # - se check_JobArea == True e PhoneNumber == False   | Troca
    # - se check_JobArea == True e PhoneNumber == True    | Nan
    # - se check_JobArea == False e PhoneNumber == False  | Nan
    # - se check_JobArea == False e PhoneNumber == True   | Nada

    for indice, linha in base_pessoas.iterrows ():
      if linha["Check_Job_Area"] == True and linha['Check_phoneNumber'] == False:
        base_auxiliar.loc[indice,"jobArea"] = linha["phoneNumber"]
        base_auxiliar.loc[indice,"phoneNumber"]= linha ["jobArea"]

      if linha["Check_Job_Area"] == True and linha['Check_phoneNumber'] == True:
        base_auxiliar.loc[indice,"jobArea"] = np.nan

      if linha["Check_Job_Area"] == False and linha['Check_phoneNumber'] == False:
        base_auxiliar.loc[indice,"phoneNumber"]=np.nan

      if linha["Check_Job_Area"] == False and linha['Check_phoneNumber'] == True:    
        pass
    return base_auxiliar


# corrigir UFs utilizando um dataframe externo para correção do nosso data frame fake

from fuzzywuzzy import fuzz

def corrige_uf(estado):
  # pegamos um dataset de ufs.cvs covencional, gerando uma matriz transposta e interamos, assim, gerando um dicionario
  UFS = {row['uf_code']:row['name'] for index,row in bases_uf [["uf_code", "name"]].T.iteritems()}
  uf_result = 0
  for uf,state in UFS.items():
    # checa se os estado (UF) contem duas letras
     if len(estado) > 2:
        # Aplicando calculo da Distância de Levenshtein, um fuzzi.radio é mais preciso, ja fuzzi.partial, pega distancia aproximada
        if fuzz.ratio(estado, state)  ==  100 and fuzz.partial_ratio(estado, state) == 100:
            uf_result = uf
     else:
        uf_result = estado    
  return uf_result
```
